```python
#3 answer
comment = "please subscribe to my channel click this now and make a lot of money."
spamlist = ["make a lot of money", "buy now", "subscribe this", "click this"]

# ---------- 1st way (using find) ----------
checking_comment = comment.lower()
result1 = False
count=0
for i in spamlist:
    if checking_comment.find(i.lower()) > -1:        
        count+=1

if(count>0):
    result1=True    

if result1:
    print(f"number of spam words matched {count}")
    print("Its a spam comment (method 1)")
else:
    print("Its a normal comment (method 1)")


# ---------- 2nd way (using 'in') ----------
result2 = False
count=0
for i in spamlist:
    if i.lower() in comment.lower():
        count+=1

if(count>0):
    result2=True     

if result2:
    print(f"number of spam words matched {count}")
    print("Its a spam comment (method 2)")
else:
    print("Its a normal comment (method 2)")
```

```python
#6 answer
# grade of a student from his marks from the 
# following scheme: 
# 90 – 100 => Ex 
# 80 – 90 => A 
# 70 – 80 => B 
# 60 – 70  =>C 
# 50 – 60 => D 
# <50        => F


student=input("Enter your name:\n")
marks=int(input("Enter your marks:\n"))
if (marks>=90 and marks<=100):
    print(f"{student} grade is: Ex")
elif (marks>=80 and marks<90):
    print(f"{student} grade is: A")
elif (marks>=70 and marks<80):
    print(f"{student} grade is: B")
elif (marks>=60 and marks<70):
    print(f"{student} grade is: C")
elif (marks>=50 and marks<60):
    print(f"{student} grade is: D")
elif (marks<50):
    print(f"{student} grade is: F")
```
