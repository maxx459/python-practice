```python
#1 answer
name=input("enter your name:")
print(f"Good Afternoon {name.capitalize()}!")
```

```python
#2 answer
Name=input("enter your name:")
Date=input("enter the date:")
letter = '''  
Dear <|Name|>, 
You are selected! 
<|Date|> 
        ''' 
print(f"Dear {Name},\nYou are selected!\n{Date}")
print(letter.replace("<|Name|>", Name).replace("<|Date|>", Date)) 
```

```python
#4 answer
import re
string=input("enter a string:")
print("removes single double space:",string.replace("  "," "))
print("removes all double spaces:",re.sub("\\s+"," ",string))
```
