```python
#1 Answer

fruits=[]
print("enter 7 fruits:\n")
for i in range(7):
    fruits.append(input(f"fruit {i+1}: "))
print(fruits)
```

```python
#2 Answer

student=[]
print("enter 6 students marks:\n")
for i in range(6):
    student.append(int(input(f"student {i+1} marks: ")))
student.sort()
print("sorted list:",student)
```
