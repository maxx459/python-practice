```python
#4 answer
from random import randint

def divisibleby5(n):
    if n%5==0:
        return True
    return False

list1=[randint(1,100) for i in range(20)]

list2=list(filter(divisibleby5,list1))

print(list2)
```

```python
#5 answer

from functools import reduce

def maxoflist(a,b):
    if a>b:
        return a
    return b

list1=[randint(1,100) for i in range(20)]
print(list1)

max=reduce(maxoflist,list1)

print("maximum:",max)
```
