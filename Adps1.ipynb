```python
#5 answer

num=int(input("Enter table num:"))
list=[f"{num} x {i} = {num*i}\n" for i in range(1,11)]
print(list)
file_path1="./files/table.txt"
with open(file_path1,"w") as f:
    f.writelines(list)
```
