```python
#1 answer 

def Big_Of_3(a,b,c):
    if(a>b and a>b and a>c):
        print(f"the greatest number is: {a}")
    elif(b>c):
        print(f"the greatest number is: {b}")
    else:
        print(f"the greatest number is: {c}")

list=[]
for i in range(3):
    list.append(int(input(f"Enter number {i}:\n")))

a,b,c=list
Big_Of_3(a,b,c)
```

```python
#4 answer

def sum_of_natural_nums(a):
    if a==0:
        return 0
    elif a==1: 
        return 1 
    return a+sum_of_natural_nums(a-1)

num=int(input("Enter number: "))
print("sum of n natural numbers:",sum_of_natural_nums(num))
```

```python
#5 answer

def pattern(a):
    for i in range(a,0,-1):
        print("*"*i)

n=int(input("Enter the length of pattern:"))
pattern(n)
```

```python
#6 answer

def inchs_cms(inches):
    cms=inches*2.54
    print(f"Inches to centimeters: {cms}")

inches=int(input("enter Inches:\n"))
inchs_cms(inches)
```

```python
#7 answer
def clean_and_remove(word_list, target):
    
    result = []
    
    for word in word_list:

        cleaned_word = word.strip()
        
        if cleaned_word != target:
            
            result.append(cleaned_word)
            
    return result

my_words = ["  pizza ", " burger", "  pizza", "taco  "]
to_delete = "pizza"

final_list = clean_and_remove(my_words, to_delete)

print(f"Final List: {final_list}")

#we can also use remove() func for revmoving a word

```
