```python
import random

def game():
    print("Game started")
    return random.randint(1, 100)

score = game()
print("Your score:", score)

# Try reading old high score
try:
    with open("./files/hi-score.txt", "r") as f:
        content = f.read()
        if content == "":
            high_score = 0
        else:
            high_score = int(content)
except FileNotFoundError:
    high_score = 0

# Compare and update
if score > high_score:
    with open("./files/hi-score.txt", "w") as f:
        f.write(str(score))
    print("New High Score Updated!")
else:
    print("High Score remains:", high_score)
```

```python
#4 answer
import re
with open("./files/donkey.txt") as f:
    paragraph=f.read() #string type

#paragraph=paragraph.replace("Donkey","#####")# use this for simple replacement of extact mactches
paragraph=re.sub("donkey","#####",paragraph,flags=re.IGNORECASE)

print(paragraph)

with open("./files/donkey1.txt","w") as f:
    f.writelines(paragraph)
```

```python
#5 answer
import re
with open("./files/donkey.txt") as f:
    paragraph=f.readlines() #list type
new_paragraph=[]
for i in paragraph:
    #i=i.replace("Donkey","#####")# use this for simple replacement of extact mactches
    i=re.sub("donkey","#####",i,flags=re.IGNORECASE)
    new_paragraph.append(i)
    print(i)
#print(new_paragraph)
with open("./files/donkey.txt","w") as f:
    f.writelines(new_paragraph)
```
