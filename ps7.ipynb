```python
#4 answer

def check_prime():
    while True:
        while True:
            try:
                num = int(input("Enter a number: "))
                break
            except Exception:
                print("Enter only numbers")
        if num < 0:
            print("Retry with positive numbers")
        else:
            break

    if num == 0 or num == 1: 
        print("Its not a prime number")
        return # Exits the function early

    for i in range(2, num):
        if num % i == 0:
            print("Its not a prime number")
            return # Exits the function early

    print("It is a prime number")

check_prime()

```

```python
#5 answer

num=int(input("Enter the number of n natural numbers you want to find sum for:\n"))
res=0
for i in range(1,num+1):
    res+=i
print("Through for loop")
print("sum:",res)
ii=1
res2=0
while(ii<=num):
    res2+=ii
    ii+=1
print("Through while loop")
print("sum:",res2)
```

```python
#6 answer

num=int(input("Enter a number:\n"))
res=1
for i in range(1,num+1):
    res*=i
print("factorial:",res)
```

```python
#7 answer

lines=int(input("Enter the lenght of the * pyramid: "))
print("one way to do it:\n")
for i in range(1,lines+1):
    spaces=" "*(lines-i)
    stars="*"*(i*2-1)
    print(spaces+stars)
print("second way to do it:\n")
for i in range(1,lines+1):
    print(" "*(lines-i),end="")
    print("*"*(i*2-1),end="")
    print("")

# spaces=" "
# stars="*"
# print(spaces*(lines-1)+stars)
```

```python
#9 answer
lines=int(input("Enter the lenght of the * square pyramid(min value-3): "))
stars="*"*(lines)
lstar="*"
spaces=" "*(lines-2)
rstar="*"

print(stars)
for i in range(1,lines-1):
    print(lstar+spaces+rstar)
print(stars)
```

```python
#10 answer

num=int(input("Enter your the table you want in reverse order:\n"))

for i in range(10,0,-1):
    print(f"{num} x {i} = {num*i}")
```
