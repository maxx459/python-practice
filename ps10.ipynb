```python
#2 answer
class calculator:
    def __init__(self,a):
        self.a=a
    @staticmethod
    def square_root(a):
        return a**0.5
    @staticmethod
    def square(a):
        return a**2
    @staticmethod
    def cube(a):
        return a**3

cal=calculator(int(input("enter a number:")))
print(f"Square of the given number: {cal.square(cal.a)}")
print(f"cube of the given number: {cal.cube(cal.a)}")
print(f"Square of the given number: {cal.square_root(cal.a):.2f}")

```

```python
#3 answer
class test:
    a = 55  # Class Attribute

    def __init__(self):
        print("Initial Class Value:", test.a)

    def printInfo(self):        
        print("--- Verification ---")
        print("Class attribute (test.a):", test.a)   # Still 55
        print("Instance attribute (self.a):", self.a) # The new value (e.g., 0)
        print("\nConclusion: The class attribute did NOT change.")
        print("Instead, a new 'a' was created specifically for this object.")

tester = test()
# This creates a 'shadow' variable on the object
tester.a = int(input("Enter a value (try 0): ")) 
tester.printInfo()
                
```

```python
#4 answer
class test:
    a = 55  # Class Attribute

    def __init__(self):
        print("Initial Class Value:", test.a)

    @staticmethod
    def greet():
        print("hello!")

    def printInfo(self):        
        print("--- Verification ---")
        print("Class attribute (test.a):", test.a)   # Still 55
        print("Instance attribute (self.a):", self.a) # The new value (e.g., 0)
        print("\nConclusion: The class attribute did NOT change.")
        print("Instead, a new 'a' was created specifically for this object.")

tester = test()
# This creates a 'shadow' variable on the object
tester.a = int(input("Enter a value (try 0): ")) 
tester.greet()
tester.printInfo()
                
```
