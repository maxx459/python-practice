```python
#1 answer

class TwoDVector:
    def __init__(self, i,j):
        self.i=i
        self.j=j
    def show(self):
        print(f"The 2-D Vector is {self.i}i + {self.j}j")
class ThreeDVector(TwoDVector):
    def __init__(self, i, j,k):
        super().__init__(i, j)
        self.k=k
    def show(self):
        print(f"The 3-D Vector is {self.i}i + {self.j}j + {self.k}k")

a=TwoDVector(1,2)
b=ThreeDVector(4,5,6)
a.show()
b.show()
```

```python
#2 answer
class animal:
    @staticmethod
    def eyes():
        print("Eyes: 2")
    @staticmethod    
    def ears():
        print("Ears: 2")
    @staticmethod    
    def legs():
        print("legs: 4")
class pet(animal):
    @staticmethod
    def collar():
        print("Have a collar")
    @staticmethod
    def house():
        print("Lives in a house")
class dog(pet):
    @staticmethod
    def bark():
        print("Dog can bark")

obj=dog()
obj.bark()
obj.house()
obj.collar()
obj.eyes()
obj.ears()
obj.legs()
    
```

```python
#3 answer
class Employee:
    salary=12000
    increament=10
    # def __init__(self,salary):
    #     self.salary=salary
        
    @property
    def salaryAfterIncrement(self):
        return f"Your current salary {self.salary}\nYour increamented salary:{self.salary+(self.salary*(self.increament/100))}"
    @salaryAfterIncrement.setter
    def salaryAfterIncrement(self,new_salary):
        self.increament= ((new_salary/self.salary)-1)*100
    
user=Employee()
# user.salaryAfterIncrement=12000
print(user.salaryAfterIncrement)
print(f"Increamented by: {user.increament}%")
```

```python
#4 answer

class Complex:
    def __init__(self, r, i):
        self.r = r
        self.i = i

    def __add__(self, other): #here c1 is self and the other is c2
        return Complex(self.r + other.r, self.i + other.i)
    
    def __mul__(self, other):
    # self = the first number (left side of *)
    # other = the second number (right side of *)
        real_part = self.r * other.r - self.i * other.i
        imag_part = self.r * other.i + self.i * other.r
        return Complex(real_part, imag_part)

    
    def __str__(self):
        return f"{self.r} + {self.i}i"
        
    

c1 = Complex(1, 2)
c2 = Complex(3, 4)
print(c1 + c2)
print(c1*c2)
```

```python
#5 answer

class Vector:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def __add__(self, other):
        result = Vector(self.x + other.x, self.y + other.y, self.z + other.z)
        return result

    def __mul__(self, other):
        result = self.x * other.x + self.y * other.y + self.z * other.z
        return result

    def __str__(self):
        return f"Vector({self.x}, {self.y}, {self.z})"

# Test the implementation
v1 = Vector(1, 2, 3)
v2 = Vector(4, 5, 6)
v3 = Vector(7, 8, 9)  # Same dimension vector

print(v1 + v2)  # Output: Vector(5, 7, 9)
print(v1 * v2)  # Output: 32

print(v1 + v3)  # Output: Vector(8, 10, 12)
print(v1 * v3)  # Output: 50

print(v1 + v3)  # Output: Vector(8, 10, 12)
print(v1 * v3)  # Output: 50
```
